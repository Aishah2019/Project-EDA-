
### Question 2:
•	Does the course price affect subscribers count ?

```python
# price does not affect on the subscribers count 
#We know this by looking at the correlation heat map (Correlation = 0.051)
plt.scatter(df['price'],df['num_subscribers'])
plt.title('Price With Subscribers Count')
plt.xlabel("Price")
plt.ylabel("num_subscribers")     
```

### Question 3:
•	Does the course duration affect the price?

```python
plt.figure(figsize=(12,12))
plt.grid()
plt.xlabel("price")
plt.ylabel("content_duration")
plt.plot(df.groupby("price")[["content_duration"]].max());
                 
```

```python
# course duration does not affect price
plt.scatter(df['price'],df['content_duration'])
```

### Question 4:
•	What is The Trends in Learning ?

### *Topics of the courses in the years*
_____________________________________________________

```python
#numbers of different subjects correspending to the year
# Business Finance is needed the last few years 
df['year']= df['published_timestamp'].dt.year
subject_and_year = df.groupby('year')['subject'].value_counts().reset_index(level=0).rename(columns={'subject': 'subject count'}, index={'index': 'Subject'})
subject_and_year
```

```python
# most of the poeple are interested in Business Finance And Web Development 
sns.countplot('subject',data=df)
```

```python
# most of the courses are cheap [25-60]
sns.histplot(data=df,x="price",kde=True);
```

```python
# most of the course a re shor time courses [1-8]
plt.figure(figsize=(10,6))
sns.histplot(data=df,x="content_duration",bins=50)
```

```python
# more subscribers leads to more reviews 
# the price does not affect the subscripers count 
# course time does not affect the course time 
sns.heatmap(df.corr(), cmap="coolwarm", annot=True);
```
