### Topics of the courses in the years

```python
#numbers of different subjects correspending to the year
# Business Finance is needed the last few years 
df['year']= df['date'].dt.year
subject_and_year = df.groupby('year')['subject'].value_counts().reset_index(level=0).rename(columns={'subject': 'subject count'}, index={'index': 'Subject'})
subject_and_year
```

### What are the first five paid courses according to subscribers?

#### Top 5 Paid Courses

```python
# the most subscribers are in web development 
top_5_paid_courses = df[df['price']!=0][['course_title','year','subject','num_subscribers']].sort_values(by= 'num_subscribers',ascending=False).head(5)
top_5_paid_courses
```
